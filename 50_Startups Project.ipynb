<h3><b>6) Create Dependent(y) and Independent(X) Variables</b></h3>

```python
target_feature = 'Profit'

# Separate object for Traget feature
y = df[target_feature]

# Separate object for Input Features
X = df.drop(target_feature, axis=1)
```

<h3><b>7) Split Dataset to Train and Test</b></h3>

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, train_size = 0.7, random_state = 1)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

<h3><b>8) Build the Model (Linear Regression)</b></h3>

   <b>Apply Linear Regression on Train Dataset</b>      

```python
from sklearn.linear_model import LinearRegression
```

```python
mlr = LinearRegression()
```

```python
mlr.fit(x_train, y_train)
```

<h3><b>9) Build the Model (Linear Regression)</b></h3>

```python
x_test.head()
```

```python
x_test.shape
```

<h3><b>10) Apply the Trained Model on Test Dataset to get the Predicted Values</b></h3>

```python
y_pred = mlr.predict(x_test)
y_pred
```

```python
y_pred.shape
```

```python
y_test
```

<h3><b>11) Compare the Actual output(y_test) Values with the Predicted values(y_pred)</b></h3>

```python
df1 = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred, 'Variance': y_test-y_pred})
```

```python
df1
```

<h3><b>12) Check the Models Accuracy</b></h3>

```python
testing_data_model_score = mlr.score(x_test, y_test)
print("Model Accuracy on Testing data",testing_data_model_score)

training_data_model_score = mlr.score(x_train, y_train)
print("Model Accuracy on Training data",training_data_model_score)
```

<h3><b>13) Check the Models R2 Score</b></h3>

```python
from sklearn.metrics import r2_score

r2Score = r2_score(y_pred, y_test)
print("R2 score of model is :" ,r2Score*100)
```
